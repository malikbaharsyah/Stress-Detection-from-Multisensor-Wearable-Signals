```python
import pandas as pd

# Load the Excel file
df = pd.read_excel(excel_path, sheet_name=None)

# Combine all sheets (participants) into one DataFrame
survey_all = pd.concat(df.values(), ignore_index=True)

# Convert to datetime using the date + time columns
survey_all["Start"] = pd.to_datetime(survey_all["date"].astype(str) + " " + survey_all["Start time"].astype(str))
survey_all["End"] = pd.to_datetime(survey_all["date"].astype(str) + " " + survey_all["End time"].astype(str))

# Convert to UNIX timestamp (seconds since epoch)
survey_all["Start_unix"] = survey_all["Start"].astype('int64') // 10**9
survey_all["End_unix"] = survey_all["End"].astype('int64') // 10**9

# Keep only useful columns
survey_labeled = survey_all[["ID", "Start", "End", "Start_unix", "End_unix", "Stress level"]]
```

```python
# delete na values
survey_labeled = survey_labeled[survey_labeled["Stress level"] != "na"]
# sort survey_labeleed by ID
# convert ID to string
survey_labeled["ID"] = survey_labeled["ID"].astype(str)
survey_labeled = survey_labeled.sort_values(by=["ID"])
```

```python
import pandas as pd

# Load CSVs
data_all = pd.read_csv("data_all.csv")

# Step 1: Build labeled_data_dict
labeled_data_dict = {}
for _, row in survey_labeled.iterrows():
    id = row["ID"]
    labeled_data_dict.setdefault(id, []).append({
        "start": row["Start_unix"],
        "end": row["End_unix"]
    })
count_labeled = sum(len(times) for times in labeled_data_dict.values())
print("Count labeled data:", count_labeled)

# Step 2: Build data_all_dict
data_all_dict = {}
for _, row in data_all.iterrows():
    filename = row["filename"]
    id, unix_time = filename.split("_")
    unix_time = int(unix_time)
    data_all_dict.setdefault(id, []).append(unix_time)

# Step 3: Count before filtering
count_all = sum(len(times) for times in data_all_dict.values())
print("Before filtering:", count_all)

# Step 4: Filter out unmatched times safely
filtered_data_all_dict = {}
for id, times in data_all_dict.items():
    if id in labeled_data_dict:
        valid_times = []
        for unix_time in times:
            if any(ld["start"] <= unix_time <= ld["end"] for ld in labeled_data_dict[id]):
                valid_times.append(unix_time)
        if valid_times:
            filtered_data_all_dict[id] = valid_times

# Step 5: Count after filtering
count_filtered = sum(len(times) for times in filtered_data_all_dict.values())
print("After filtering:", count_filtered)

data_all_dict
```
